```python
#preprocessing

n_cols = 1
dataset = abc['close']
dataset = pd.DataFrame(dataset)
data = dataset.values

data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
scaled_data = scaler.fit_transform(np.array(data))
```

```python
train_size = int(len(data) * 0.75)
test_size = len(data) - train_size

print('Train Size:', train_size, 'Test Size:', test_size)
```

```python
train_data = scaled_data[0:train_size, :]
train_data.shape
```

```python
x_train = []
y_train = []
time_steps = 60
n_cols = 1

for i in range(time_steps, len(train_data)):
    x_train.append(train_data[i-time_steps:i, :n_cols])
    y_train.append(train_data[i, :n_cols])
    
    if i <=time_steps:
        print('X_train:', x_train)
        print('y_train:', y_train)
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],n_cols))
```

```python
x_train.shape, y_train.shape
```

```python
#lstm model
```

```python
model = Sequential([
    LSTM(50, return_sequences = True, input_shape=(x_train.shape[1], n_cols)),
    LSTM(64, return_sequences = False),
    Dense(32),
    Dense(16),
    Dense(n_cols)
])

model.compile(optimizer = 'adam', loss= 'mse', metrics = 'mean_absolute_error')
```

```python
model.summary()
```

```python
# fitting lstm model to training set
```

```python
history = model.fit(x_train, y_train, epochs = 20, batch_size= 32)
```

```python
# model eveluation

plt.figure(figsize=(12,8))
plt.plot(history.history['loss'])
plt.plot(history.history['mean_absolute_error'])
plt.legend(['Mean Squared Error', 'Mean Absolute Error'])
plt.title("Losses")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.show()
```

```python
#prediction

time_steps = 60
test_data = scaled_data[train_size - time_steps:, :]

x_test = []
y_test = []
n_cols = 1

for i in range(time_steps, len(test_data)):
    x_test.append(test_data[i - time_steps:i, 0:n_cols])
    y_test.append(test_data[i, 0:n_cols])
    
x_test, y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], n_cols))    
```

```python
x_test.shape, y_test.shape
```

```python
prediction = model.predict(x_test)
```

```python
prediction = scaler.inverse_transform(prediction)
prediction.shape
```

```python
y_test = scaler.inverse_transform(y_test)
```

```python
rmse = np.sqrt(np.mean(y_test - prediction)**2).round(2)
rmse
```

```python
preds_acts = pd.DataFrame(data={'Predictions' : prediction.flatten(), 'Actuals' : y_test.flatten()})
preds_acts
```

```python
plt.figure(figsize=(16,6))
plt.plot(preds_acts['Predictions'])
plt.plot(preds_acts['Actuals'])
plt.legend(['Predictions','Actuals'])
plt.show()
```

```python
train = dataset.iloc[:train_size]
test = dataset.iloc[train_size:]
test = test.copy()
test['Predictions'] = prediction

plt.figure(figsize=(16,6))
plt.title("Bitcoin stock price predictions", fontsize=20)
plt.xlabel("Date", fontsize=18)
plt.ylabel("Close Price", fontsize=18)
plt.plot(train['close'], linewidth=3)
plt.plot(test['close'], linewidth=3)

plt.plot(test['Predictions'], linewidth=3, color='green')
plt.legend(['Train', 'Test', 'Prediction'])
```

```python
# Predict the next N days
N = 60  # Number of days to predict
last_date = abc['date'].max()  # Get the last date directly
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=N, closed='left')

future_preds = []

last_60_days = scaled_data[-time_steps:]

for i in range(N):
    input_data = last_60_days[-time_steps:]
    input_data = np.array(input_data)
    input_data = np.reshape(input_data, (1, time_steps, n_cols))

    predicted_price = model.predict(input_data)
    future_preds.append(predicted_price[0][0])

    last_60_days = np.append(last_60_days, predicted_price, axis=0)

future_preds = scaler.inverse_transform(np.array(future_preds).reshape(-1, 1))

future_price_df = pd.DataFrame(future_preds, columns=['Predicted Close Price'], index=future_dates)
future_price_df
```
